```python
import os
import numpy as np
import cv2
from os.path import join as pjoin

def useMaskImage(mask, image):
    mask = np.where((mask==0),0,1).astype('uint8')
    return mask*image


masout = []

for name in os.listdir('nails/images'):
    if (os.path.isfile(name)):
        continue
    image = cv2.imread(pjoin('nails/images', name))
    mask  = cv2.imread(pjoin('nails/labels', name))
    masout.append(useMaskImage(mask, image))

idx = 0
while True:
    idxtext = cv2.putText(masout[idx],'a/d', (0, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)
    cv2.imshow("Nails", idxtext)
    
    key = cv2.waitKey(0) & 0xff
    if key == ord('a') or key == ord('A'):
        idx = (idx - 1) % len(masout)
    if key == ord('d') or key == ord('D'):
        idx = (idx + 1) % len(masout)
    if key == 27:
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import time
import asyncio

cap = cv2.VideoCapture(0)


cap.set(3,1280) # установка размера окна

timer = time.perf_counter()
while True:
    
    ret, frame1 = cap.read()
    ret, frame2 = cap.read()
    diff = cv2.absdiff(frame1, frame2)
    
    gray = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY) # перевод кадров в черно-белую градацию
    blur = cv2.GaussianBlur(gray, (5, 5), 0) # фильтрация лишних контуров 
    _, thresh = cv2.threshold(blur, 20, 255, cv2.THRESH_BINARY) # метод для выделения кромки объекта белым цветом
    dilated = cv2.dilate(thresh, None, iterations = 3) 
    сontours, _ = cv2.findContours(dilated, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE) # нахождение массива контурных точек


    image = np.zeros([300, 300, 3], dtype=np.uint8)
    image[:300, :300] = [0, 255, 0]
    

   # for contour in сontours:
    if time.perf_counter() - timer > 3:
        if сontours!=0:
            cv2.drawContours(frame1, сontours, -1, (0, 0, 255), 2) 
            image[:300, :300] = [0, 0, 255]
    if time.perf_counter() - timer > 6:
        timer = time.perf_counter()

               
    if 6>time.perf_counter() - timer > 3:
        cv2.putText(frame1, 'Red light', (10, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
    if time.perf_counter() - timer > 6 or time.perf_counter() - timer < 3:
        cv2.putText(frame1, 'Green light', (10, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow('Moving', image)
    
    cv2.imshow("Game", frame1)
    frame1 = frame2  #
    ret, frame2 = cap.read() #  

    key = cv2.waitKey(20)
    
    if (key == ord('q')) or key == 27:
        break

        

cap.release()
cv2.destroyAllWindows()
```
